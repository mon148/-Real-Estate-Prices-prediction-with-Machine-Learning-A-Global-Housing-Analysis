#### Checking which of the columns still ahve the highest missing values

```python
imp_cols = [
    "building_total_floors", "apartment_floor", 
    "apartment_rooms","apartment_bedrooms",
    "apartment_bathrooms","building_construction_year"
]

for col in imp_cols:
    # Ensure correct variable name
    median_val = all_data[col].median()*100
    print(f"missing in {col} {median_val} ")
```

**Since the majority have the highest missing values, it would be best if the most important ones that would be very useful for the EDA and modelling to be filtered out and consider dropping other having above 60%**

```python
imp_cols = [
    "building_total_floors", "apartment_floor", 
    "apartment_rooms","apartment_bedrooms",
    "apartment_bathrooms", "price_in_USD", "building_construction_year"
]

for col in imp_cols:
    # Ensure correct variable name
    median_val = all_data[col].median()                   # calculate median
    all_data[f"{col}_missing"] = all_data[col].isna().astype(int)  # create missing flag
    all_data[col] = all_data[col].fillna(median_val)    # fill NaNs with median
    
    # Optional: print how many NaNs remain
    print(f"{col} missing after fill: {all_data[col].isna().sum()}")
```

**Checking the percentages of the columns that are important and that will beuseful for EDA and modelling**

```python
each_column = [
    "country", "location",
    "price_in_USD","apartment_total_area"
]
for cols in each_column:
    perc_missing = all_data[cols].isna().mean()*100
    print(f"percentage of missing values in {cols}: {perc_missing}")
```

**Since the percentages aren't high, dropping the rows with the nan will not affect the model in a negative way**

```python
all_data.isna().sum()
```

**The missing values in the "apartment_living_area", "apartment_total_area", could not be reassigned / filled with the mean because of the string values attached to the values in the rows (m^2). So, for the missing valuies to be re-assigned or refill with the mean, the 'm^2' must be extracted.**

```python
pattern = r'(\d+)\s*(?:m²|m\^2|m2|sqm)'

all_data["apartment_total_area_m2"] = all_data["apartment_total_area"].str.extract(pattern).astype(float)
all_data["apartment_living_area_m2"] = all_data["apartment_living_area"].str.extract(pattern).astype(float)

all_data.isna().sum()
```

**Since the two columns are all float, and the m2 has been assigned to the columns' heading, then the missing values can be attended to now** 

```python
Apartment_nan = [
    "apartment_total_area_m2",
    "apartment_living_area_m2"
]

for col in Apartment_nan:
    median_values = all_data[col].median()*100
    all_data[col] = all_data[col].fillna(median_values)  
```

**Since new columns 'apartment_total_area_m2, apartment_living_area_m2' had been extracted from the other coluns 'apartment_living_area, apartment_total_area', I consider dropping the old columns**

```python
all_data.drop(columns=["apartment_living_area"], inplace=True)

all_data.columns
```

**Since the rest have little NAN, then, dropping them won't negatively affect the EDA and the modelling**

**Working on extracting some columns from the 'title' column using regex so as to finally drop the 'title' column**

```python
# Make a copy to avoid Setting With Copy Warning
df = all_data.copy()

# 3. Extract area like “85 m^2”, “120sqm”, “200 m2”
df["area_from_title m\u00b2"] = df["title"].str.extract(
    r'(\d+)\s*(?:m2|m²|sqm|sq\.m|square meters?)',
    flags=re.IGNORECASE
).dropna()

# 5. Extract simple property type and others that could stand as columns on their own
df["property_type"] = df["title"].str.extract(
    r'(apartment|villa|studio|duplex|penthouse|house)',
    flags=re.IGNORECASE
)
```
